 1) What is hypothesis testing in statistics?
 ans. Hypothesis testing is a structured method to evaluate assumptions (hypotheses) about a population parameter using sample data. It helps determine whether the observed data supports a specific claim or if the differences are due to random chance

 2)  What is the null hypothesis, and how does it differ from the alternative hypothesis?
 ans. Null Hypothesis:
        - The null hypothesis is the default assumption that there is no effect, no difference, or no relationship between variables
      Alternate Hypothesis:
        - The alternative hypothesis is what you want to prove or support. It suggests that there is an effect, difference, or relationship.

3) Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
ans. The significance level in hypothesis testing is the probability of rejecting the null hypothesis (H₀) when it is actually true. It represents the maximum acceptable risk of making a Type I error.

4) What are Type I and Type II errors? Give examples of each.?
ans. Type I:- Rejecting the null hypothesis (H₀) when it is actually true.
ex:- H₀: The drug has no effect.
- You reject H₀ and claim the drug works.
- But in reality, the drug doesn't work.

Type II:-Failing to reject the null hypothesis when it is actually false
ex:-- H₀: The drug has no effect.
- You fail to reject H₀ and say the drug doesn’t work.
- But in reality, the drug does work.

5)  What is the difference between a Z-test and a T-test? Explain when to use each?
ans. Z-test:
      - population standard deviation is known
      - sample size is greater than 30
     T-test:
      - population standard deviation is not known
      - sample size is smaller than 30

```python
#8 Write a Python script to simulate data from a normal distribution and
#calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
from scipy import stats
np.random.seed(42)
mu=100
sigma=15
n=200
data=np.random.normal(loc=mu,scale=sigma,size=n)
sample_mean=np.mean(data)
standard_error=stats.sem(data)
confidence_level = 0.95
ci = stats.norm.interval(confidence_level, loc=sample_mean, scale=standard_error)
plt.hist(data)
plt.show()
```

```python
#9 Write a Python function to calculate the Z-scores from a dataset and
#visualize the standardized data using a histogram. Explain what the Z-scores represent
#in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = [(x - mean) / std_dev for x in data]
    plt.hist(z_scores)
    plt.xlabel('Z-Score (Standard Deviations from Mean)')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

    return z_scores
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores = calculate_and_plot_z_scores(sample_data)

```

 1) What is a random variable in probability theory?
 Ans. A random variable is a function that maps outcomes from a sample space to real numbers.

2) What are the types of random variables?
Ans. 1. Discrete Random Variable
        - Takes on countable values (like integers).
        - Example: Number of heads when flipping 3 coins. Possible values: 0, 1, 2, 3.
     2. Continuous Random Variable
        - Takes on uncountably infinite values (like any value in an interval).
        - Example: Time it takes for a bus to arrive, which could be any value between 0 and 30 minutes.

3) Explain the difference between discrete and continuous distributions.
ans. A discrete distribution describes the probability of outcomes of a discrete random variable, which can take on countable values (like whole numbers).
A continuous distribution describes the probability of outcomes of a continuous random variable, which can take on any value within a range.

4) What is a binomial distribution, and how is it used in probability?
Ans A binomial distribution models the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes that is success or failure
It is useful in:
- Helps calculate probabilities in yes/no scenarios.
- Used in quality control, clinical trials, marketing, and machine learning.
- Forms the basis for hypothesis testing and confidence intervals in statistics.

5) What is the standard normal distribution, and why is it important?
Ans. The standard normal distribution is a normal distribution with:
- Mean (μ) = 0
- Standard deviation (σ) = 1

It is useful for statistical inference and probability calculation

6)What is the Central Limit Theorem (CLT), and why is it critical in statistics?
ans. When you take a large number of independent, random samples from any population (with a finite mean and variance), the distribution of the sample means will approximate a normal distribution — regardless of the original population's shape.
It simplifies complex problems and is widely used in machine learning

7) What is the significance of confidence intervals in statistical analysis?
Ans. A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (like a mean or proportion) with a certain level of confidence.
 significance:
 - used omn hypothesis testing
 -supports decision making

8) What is the concept of expected value in a probability distribution?
ans.The expected value is like the long-term average or the center of gravity of a probability distribution. It tells you what outcome you can expect on average if you repeat an experiment many times.






```python
#: Write a Python program to generate 1000 random numbers from a normal
#distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
#deviation using NumPy, and draw a histogram to visualize the distribution.
import numpy as np
import matplotlib.pyplot as plt
mean = 50
std_dev = 5
sample_size = 1000

data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

computed_mean = np.mean(data)
computed_std = np.std(data)

print(f"Computed Mean: {computed_mean:.2f}")
print(f"Computed Standard Deviation: {computed_std:.2f}")

plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
